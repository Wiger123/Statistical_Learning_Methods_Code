```
import cv2
import os
import csv
import numpy as np
from google.colab import files
```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
# 图像矩阵
pic = []
```

```
with open('/content/drive/MyDrive/csv_data/flash_data_8.csv', 'r') as csvfile:
  # 生成一个 csv_reader 用于迭代读出
  csv_reader = csv.reader(csvfile)

  # 跳过前几行
  next(csv_reader)

  # 行列数
  m, n = 0, 0

  # 读取像素
  for i in csv_reader:
    m += 1
    for j in i[:-3]:
      n += 1
      pic.append(j)
  # 行数
  n = n // m
```

```
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
```

```
# 显示图片
pic = np.array(pic)
pic = np.resize(pic, (m, n))
print(pic)
# 显示图片
plt.imshow(pic.astype(np.float32))
plt.show()
```

```
# 保存图片
cv2.imwrite("test.png", pic.astype(np.float32))

# 下载图片
files.download("test.png")
```
